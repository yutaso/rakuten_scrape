```python
from selenium import webdriver
from time import sleep
import time
import pandas as pd
from selenium.webdriver.common.by import By
```

```python
def search_premium(url):
    try:
        #driver = webdriver.Chrome('chromedriver')
        driver = webdriver.Chrome()
        driver.get(url)
        time.sleep(2)
        #search_bar = driver.find_element_by_class_name("tag-wrapper")
        search_bar = driver.find_element(By.CLASS_NAME,"tag-wrapper")
        #username = driver.find_element(By.ID, "username")
        rth_name_area = driver.find_element(By.ID,'RthNameArea')
        # 'RthNameArea'内のclassが'tag-wrapper'の要素を取得
        tag_wrapper = rth_name_area.find_element(By.CLASS_NAME,'tag-wrapper')
        # 'tag-wrapper'内のclassが'premium_icon'の要素を取得
        premium_icon = tag_wrapper.find_element(By.CLASS_NAME,'premium_icon')
        # 'premium_icon'内のimg要素を取得
        img_element = premium_icon.find_element(By.TAG_NAME,'img')
        # img要素からsrc属性の値を取得
        src_value = img_element.get_attribute('src')
        # src属性の値から'premium_ultra.svg'を抽出
        file_name = src_value.split('/')[-1]
        #print(f'抽出されたファイル名: {file_name}')
        driver.close()
    except Exception as e:
        # エラーが発生した場合（premium_iconが見つからなかった場合）
        #print(f'エラー: {e}')
        file_name = 'no'
        #print(file_name)
        driver.close()
    return file_name
```

```python
#入力ファイルをここで指定
df=pd.read_csv('over100.csv', index_col=None)
```

```python
#お試し用
df['premium_rank']=str(0)
for i in range(10):
    hotel_id = df['施設ID'][i]
    print(hotel_id)
    premium=search_premium('https://travel.rakuten.co.jp/HOTEL/'+str(hotel_id)+'/'+str(hotel_id)+'.html')
    print(premium)
    df['premium_rank'][i]=premium
```
